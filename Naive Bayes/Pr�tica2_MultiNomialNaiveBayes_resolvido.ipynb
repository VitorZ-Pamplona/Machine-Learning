**Atividade 1:**

1. Banco de email rotulados disponível em:  https://spamassassin.apache.org/old/publiccorpus/ (a base de dados foi adaptada para um arquivo único contendo apenas o assunto dos emails)
2. Carregar os dados do arquivo *spam.txt* nos dados de entrada do algoritmo de aprendizado.

    

```
def load_emails(filename):
    M = []
    y = []
    with open(filename, 'r', errors='ignore') as file:
            for line in file:
                pos = line.rfind(':')
                M.append(line[:pos])

                label = line[pos+1:].strip()
                y.append(True if (label == "True") else False)

    return M,y
                    
M, y = load_emails("spam.txt")
for msg, _y in zip(M, y):
    print(msg + " : " + str(_y))
```

**Atividade 2:**

1. Implementar um separador de palavras das mensagens (tokenize)


```
import re

def tokenize(message):
    message = message.lower() #minusculo
    all_words = re.findall("[a-z0-9']+", message) #separa as palavras
    return set(all_words) #remove repeticao

print(M[10])
print(tokenize(M[10]))
```

**Atividade 3:**

1. Criar para cada mensagem um vetor de frequencias de cada palavra do dicioário deste experimento;
2. Utilizar a classe *DictVectorizer* do pacote *sklearn.feature_extraction*

```
from collections import defaultdict
from sklearn.feature_extraction import DictVectorizer

def get_matrix_count_words(M):
    counts = []
    for message in M:
        ##dicionario formado por [word][word_count]
        c = defaultdict(int)
        for word in tokenize(message):
            c[word] += 1
        counts.append(c)
  
    #Organiza os dicionários em uma matriz de contagem
    dv = DictVectorizer(sparse=False)
    return dv.fit_transform(counts)


X = get_matrix_count_words(M)

print(X)
```

**Atividade 4:**

1. Dividir entrada em dados de treino e teste.

```
from sklearn.model_selection import train_test_split

#divide os dados em teste e treino
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

**Atividade 5:**

1. Utilizar a classe *MultinomialNB* do pacote *sklearn.naive_bayes* para ajustar com os dados de treino;
2. Usar os dados de teste para computar as probabilidades e suas predições de classificação;
3. Computar as métricas de aprendizado com os dados de teste.

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

mnb = MultinomialNB()
mnb.fit(x_train, y_train)

y_prob = mnb.predict_proba(x_test)
#print(y_prob)
y_pred = mnb.predict(x_test)
#print(y_pred)

print(classification_report(y_test, y_pred))
```
